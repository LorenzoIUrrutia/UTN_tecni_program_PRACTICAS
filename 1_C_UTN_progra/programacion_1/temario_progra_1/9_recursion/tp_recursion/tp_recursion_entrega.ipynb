```python
# 1)

"""
Crea una función recursiva que calcule el factorial de un número.
Luego calcula y muestra el factorial de todos los números entre 1 y n.
"""

def factorial(n):
    """
    Calcula el factorial de n de forma recursiva.
    Factorial de n = n * (n-1) * (n-2) * ... * 1
    """
    # Caso base: factorial de 0 y 1 es 1
    if n == 0 or n == 1:
        return 1
    
    # Caso recursivo: n * factorial(n-1)
    return n * factorial(n - 1)


def mostrar_factoriales():
    """
    Muestra los factoriales desde 1 hasta el número indicado por el usuario.
    """
    numero = int(input("Ingrese un número entero positivo: "))
    
    print(f"\nFactoriales desde 1 hasta {numero}:")
    print("-" * 40)
    
    i = 1
    while i <= numero:
        resultado = factorial(i)
        print(f"Factorial de {i} = {resultado}")
        i = i + 1


# Prueba del ejercicio 1
print("=" * 50)
print("EJERCICIO 1: FACTORIAL")
print("=" * 50)
mostrar_factoriales()
print()
```

```python
#2)

"""
Crea una función recursiva que calcule el valor de Fibonacci en una posición.
Luego muestra la serie completa hasta la posición indicada por el usuario.
"""

def fibonacci(n):
    """
    Calcula el valor de Fibonacci en la posición n.
    Fibonacci: 0, 1, 1, 2, 3, 5, 8, 13, 21...
    """
    # Caso base: primeras dos posiciones
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    # Caso recursivo: suma de los dos anteriores
    return fibonacci(n - 1) + fibonacci(n - 2)


def mostrar_fibonacci():
    """
    Muestra la serie de Fibonacci hasta la posición indicada.
    """
    posicion = int(input("Ingrese la posición final de Fibonacci: "))
    
    print(f"\nSerie de Fibonacci hasta posición {posicion}:")
    print("-" * 40)
    
    i = 0
    while i <= posicion:
        resultado = fibonacci(i)
        print(f"Fibonacci({i}) = {resultado}")
        i = i + 1


# Prueba del ejercicio 2
print("=" * 50)
print("EJERCICIO 2: FIBONACCI")
print("=" * 50)
mostrar_fibonacci()
print()
```

```python
#3)

"""
Crea una función recursiva que calcule n^m usando la fórmula:
n^m = n * n^(m-1)
"""

def potencia(base, exponente):
    """
    Calcula base^exponente de forma recursiva.
    """
    # Caso base: cualquier número elevado a 0 es 1
    if exponente == 0:
        return 1
    
    # Caso recursivo: base * base^(exponente-1)
    return base * potencia(base, exponente - 1)


def calcular_potencia():
    """
    Calcula la potencia de un número según entrada del usuario.
    """
    base = int(input("Ingrese la base: "))
    exponente = int(input("Ingrese el exponente: "))
    
    resultado = potencia(base, exponente)
    print(f"\n{base}^{exponente} = {resultado}")


# Prueba del ejercicio 3
print("=" * 50)
print("EJERCICIO 3: POTENCIA")
print("=" * 50)
calcular_potencia()
print()
```

```python
#4)

"""
Función recursiva que convierte un número decimal a binario.
Método: dividir por 2 repetidamente y guardar los restos.
"""

def decimal_a_binario(n):
    """
    Convierte un número decimal a su representación binaria.
    Retorna una cadena de texto con el binario.
    """
    # Caso base: si el número es 0 o 1, retornar como string
    if n == 0:
        return "0"
    if n == 1:
        return "1"
    
    # Caso recursivo: dividir por 2 y concatenar el resto
    # El resto se obtiene con % y el cociente con //
    resto = n % 2
    cociente = n // 2
    
    # Llamada recursiva con el cociente + resto actual
    return decimal_a_binario(cociente) + str(resto)


def convertir_a_binario():
    """
    Convierte un número decimal ingresado por el usuario a binario.
    """
    numero = int(input("Ingrese un número decimal positivo: "))
    
    binario = decimal_a_binario(numero)
    print(f"\n{numero} en binario es: {binario}")


# Prueba del ejercicio 4
print("=" * 50)
print("EJERCICIO 4: DECIMAL A BINARIO")
print("=" * 50)
convertir_a_binario()
print()
```

```python
#5)

"""
Función recursiva que verifica si una palabra es palíndromo.
Sin usar [::-1] ni reversed().
"""

def es_palindromo(palabra):
    """
    Verifica si una palabra es palíndromo de forma recursiva.
    Palíndromo: se lee igual de izquierda a derecha y viceversa.
    """
    # Caso base: palabra vacía o de un carácter es palíndromo
    if len(palabra) <= 1:
        return True
    
    # Caso recursivo: comparar primer y último carácter
    if palabra[0] != palabra[len(palabra) - 1]:
        return False
    
    # Si coinciden, verificar el resto de la palabra (sin primero y último)
    palabra_medio = palabra[1:len(palabra) - 1]
    return es_palindromo(palabra_medio)


def verificar_palindromo():
    """
    Verifica si la palabra ingresada es un palíndromo.
    """
    palabra = input("Ingrese una palabra (sin espacios ni tildes): ")
    
    if es_palindromo(palabra):
        print(f"\n'{palabra}' ES un palíndromo")
    else:
        print(f"\n'{palabra}' NO es un palíndromo")


# Prueba del ejercicio 5
print("=" * 50)
print("EJERCICIO 5: PALÍNDROMO")
print("=" * 50)
verificar_palindromo()
print()
```

```python
#6) 

"""
Función recursiva que suma todos los dígitos de un número.
Sin convertir a string, solo operaciones matemáticas.
"""

def suma_digitos(n):
    """
    Suma todos los dígitos de un número de forma recursiva.
    Usa operaciones matemáticas: % (módulo) y // (división entera).
    """
    # Caso base: si el número es menor a 10, es un solo dígito
    if n < 10:
        return n
    
    # Caso recursivo: último dígito + suma del resto
    ultimo_digito = n % 10  # Obtiene el último dígito
    resto = n // 10         # Obtiene el número sin el último dígito
    
    return ultimo_digito + suma_digitos(resto)


def calcular_suma_digitos():
    """
    Calcula la suma de dígitos de un número ingresado.
    """
    numero = int(input("Ingrese un número entero positivo: "))
    
    resultado = suma_digitos(numero)
    print(f"\nLa suma de los dígitos de {numero} es: {resultado}")


# Prueba del ejercicio 6
print("=" * 50)
print("EJERCICIO 6: SUMA DE DÍGITOS")
print("=" * 50)
calcular_suma_digitos()
print()
```

```python
#7)

"""
Función recursiva que cuenta cuántos bloques necesita una pirámide.
En el nivel más bajo hay n bloques, luego n-1, n-2, ... hasta 1.
"""

def contar_bloques(n):
    """
    Cuenta el total de bloques necesarios para construir una pirámide
    que tiene n bloques en el nivel más bajo.
    """
    # Caso base: si hay 1 bloque en el nivel, solo necesita 1
    if n == 1:
        return 1
    
    # Caso recursivo: bloques del nivel actual + bloques de niveles superiores
    return n + contar_bloques(n - 1)


def calcular_bloques_piramide():
    """
    Calcula cuántos bloques necesita una pirámide según la base.
    """
    bloques_base = int(input("Ingrese cantidad de bloques en el nivel más bajo: "))
    
    total = contar_bloques(bloques_base)
    print(f"\nPara una pirámide con {bloques_base} bloques en la base")
    print(f"se necesitan {total} bloques en total")


# Prueba del ejercicio 7
print("=" * 50)
print("EJERCICIO 7: PIRÁMIDE DE BLOQUES")
print("=" * 50)
calcular_bloques_piramide()
print()
```

```python
#8)

"""
Función recursiva que cuenta cuántas veces aparece un dígito en un número.
"""

def contar_digito(numero, digito):
    """
    Cuenta cuántas veces aparece un dígito específico dentro de un número.
    """
    # Caso base: si el número es 0, no hay más dígitos que revisar
    if numero == 0:
        return 0
    
    # Obtener el último dígito del número
    ultimo_digito = numero % 10
    resto = numero // 10
    
    # Caso recursivo: contar si el último dígito coincide + revisar el resto
    if ultimo_digito == digito:
        return 1 + contar_digito(resto, digito)
    else:
        return contar_digito(resto, digito)


def buscar_digito():
    """
    Busca cuántas veces aparece un dígito en un número.
    """
    numero = int(input("Ingrese un número entero positivo: "))
    digito = int(input("Ingrese el dígito a buscar (0-9): "))
    
    cantidad = contar_digito(numero, digito)
    print(f"\nEl dígito {digito} aparece {cantidad} veces en {numero}")


# Prueba del ejercicio 8
print("=" * 50)
print("EJERCICIO 8: CONTAR DÍGITO")
print("=" * 50)
buscar_digito()
print()
```
