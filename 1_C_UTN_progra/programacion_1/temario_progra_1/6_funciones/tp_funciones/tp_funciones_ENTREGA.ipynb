```python
# Ejercicio 1: Tabla de Multiplicar

num = int(input("Ingrese un número: "))

def tabla_multiplicar(num):
    lista = []
    for i in range(1, 11):  # Range from 1 to 10
        lista.append(num * i)
    return lista

print(f"""Tabla de multiplicar del {num}
      {tabla_multiplicar(num)}"""
)
```

```python
# Ejercicio 2: Suma de números Pares

lista = []
while len(lista) < 10: # creacion de una lista manualmente, longitud establecida
    num = int(input("Ingrese los numeros de la lista: "))
    lista.append(num)
print(f"Tu lista es: {lista}")

def suma_pares(numeros):
    """ 
    % 2 descartando asi los != 0 y filtrando asi los pares
    almacena en una lista nueva los pares filtrados, los muestra
    suma los pares y da el resultado de la suma
    """
    suma = 0
    new_list = []
    for i in numeros:
        if i % 2 == 0:  # Check if the NUMBER is even, not the list2
            suma += i
            new_list.append(i)
    print("Tu lista contiene los siguientes numeros pares: ", new_list)
    return suma


resultado = suma_pares(lista)
print(f"La suma de los números pares es: {resultado}")
```

```python
# Ejercicio 3: Área y Perímetro rectángulo

longitud = int(input("Ingrese la longitud: "))
anchura = int(input("Ingrese la anchura: "))

def rectangulo(longitud, anchura):
    """Returns area and perimeter as a tuple"""
    area = longitud * anchura
    perimetro = 2 * (longitud + anchura)
    return area, perimetro  # This creates and returns a TUPLE!

area, perimetro = rectangulo(longitud, anchura)

print(f"El area y el perimetro del rectangulo son, respectivamente: {rectangulo(longitud, anchura)}")
```

```python
# Ejercicio 4:  Conversion de temperatura

temperatura_celsius = float(input("Ingrese la temperatura en grados Celsius: "))

def convertir_a_fahrenheit(celsius):
    """Convierte Celsius a Fahrenheit"""
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

temperatura_fahrenheit = convertir_a_fahrenheit(temperatura_celsius)
print(f"La temperatura en grados Fahrenheit es: {temperatura_fahrenheit}")
```

```python
# Ejercicio 5: Verificar numeros primos 

num = int(input("Ingrese un número para verificar si es primo: "))

def es_primo(num):
    """ 
    Excluye al 1 y los negativos y verifica si un número es primo 
    """
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
if es_primo(num):
    print(f"{num} es un número primo.")
else:
    print(f"{num} no es un número primo.")
```

```python
# Ejercicio 6: Calcular promedios

notas = []
while len(notas) < 5:
    nota = float(input("Ingrese la nota (0-10): "))
    if 0 <= nota <= 10:
        notas.append(nota)
    else:
        print("Nota inválida. Debe ser entre 0 y 10. Intente de nuevo.")

def calcular_promedio(lista_notas):
    """Calcula el promedio de una lista de notas"""
    if len(lista_notas) == 0:
        return 0
    suma = sum(lista_notas)
    promedio = suma / len(lista_notas)
    return promedio 

promedio = calcular_promedio(notas)
print("Las notas ingresadas son:", notas)
print(f"El promedio de las notas es: {promedio}")
```

```python
# Ejercicio 7: Factorial con validacion

num = int(input("Ingrese un número para calcular su factorial: "))

def validar_entrada(num):
    if num < 0:
        print("Por favor, ingrese un número entero no negativo.")
        return False
    return True

if validar_entrada(num):
    def factorial(n):
        if n == 0:
            return 1
        else:
            return n * factorial(n-1)

    resultado = factorial(num)
    print(f"El factorial de {num} es: {resultado}")
```

```python
# Ejercico 8: Numeros primos con funciones auxiliares

num = int(input("Ingrese el numero que desea verificar si es primo: "))

def es_divisible(n, divisor):
    return n % divisor == 0

def es_primo(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if es_divisible(n, i):
            return False
    return True

if es_primo(num):
    print(f"El numero {num} es primo.")
else:
    print(f"El numero {num} no es primo.")
```

```python
# Ejercicio 9: Conversor de temperatura con menú

celcius = float(input("Ingrese la temperatura en grados Celsius a convertir: "))

def celsius_a_fahrenheit(c):
    return (c * 9/5) + 32

def celsius_a_kelvin(c):
    return c + 273.15

print("Seleccione la conversión deseada:")
print("1. Celsius a Fahrenheit")
print("2. Celsius a Kelvin")

opcion = input("Ingrese 1 o 2: ")

# Validate input until we get a valid option
while opcion not in ['1', '2']:
    opcion = input("Opción inválida. Por favor, ingrese 1 o 2: ")

if opcion == '1':
    resultado = celsius_a_fahrenheit(celcius)
    print(f"{celcius} grados Celsius son {resultado:.2f} grados Fahrenheit.")
elif opcion == '2':  
    resultado = celsius_a_kelvin(celcius)
    print(f"{celcius} grados Celsius son {resultado:.2f} grados Kelvin.")
```
