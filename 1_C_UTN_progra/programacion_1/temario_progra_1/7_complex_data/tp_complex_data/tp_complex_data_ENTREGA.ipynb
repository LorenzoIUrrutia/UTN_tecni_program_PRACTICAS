```python
# 4) Agenda de contactos

contactos = {}

for i in range(5):
    input_nombre = input("Ingrese el nombre del contacto: ")
    input_telefono = input("Ingrese el teléfono del contacto: ")
    contactos[input_nombre] = input_telefono


print("Agenda de contactos:")
for nombre in contactos:
    print(f"{nombre}")

while True:
    nombre_seleccionado = input("Seleccione el nombre de contacto para ver su teléfono:")
    if nombre_seleccionado in contactos:
        print(f"Teléfono de {nombre_seleccionado}: {contactos[nombre_seleccionado]}")
    else:
        print("Contacto no encontrado.")
        
    continuar = input("¿Desea buscar otro contacto? (s/n): ")
    if continuar.lower() != 's':
        break
    else:
        continue
```

```python
# 5) Solicita al usuario una frase e imprime:
print("Ingrese una frase:")
frase = input()
print(f"La frase ingresada es: {frase}")
frase = frase.split()

print("Palabras únicas en la frase:")
print(set(frase))

word_count = {}
for word in frase:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print("Cantidad de veces que aparece cada palabra:")
print(word_count)
```

```python
# 6) Permití ingresar los nombres de 3 alumnos, y para cada uno una tupla de 3 notas. 
# Luego, mostrá el promedio de cada alumno

alumnos = {}
for i in range(3):
    nombre = input(f"Ingrese el nombre del alumno {i+1}: ")
    notas = tuple(float(input(f"Ingrese la nota {j+1} de {nombre}: ")) for j in range(3))
    alumnos[nombre] = notas

print("Promedio de cada alumno:")
for nombre, notas in alumnos.items():
    promedio = sum(notas) / len(notas)
    print(f"{nombre}: {promedio}")
```

```python
"""
8) Armá un diccionario donde las claves sean nombres de productos y los valores su stock. 
Permití al usuario: 
• Consultar el stock de un producto ingresado. 
• Agregar unidades al stock si el producto ya existe. 
• Agregar un nuevo producto si no existe.
"""

productos_stock = {
    'Manzanas': 50,
    'Peras': 30,
    'Naranjas': 20
}

while True:
    print("\nOpciones:")
    print("1. Consultar stock")
    print("2. Agregar unidades al stock")
    print("3. Agregar nuevo producto")
    print("4. Salir")

    opcion = input("Selecciona una opción: ")

    if opcion == "1":
        print("\nProductos y sus stocks:")
        for i in productos_stock:
            print(f"{i}: {productos_stock[i]} unidades")

    elif opcion == "2":
        print("\nAgregar unidades al stock:")
        producto = input("Ingresa el nombre del producto: ")
        unidades = int(input("Ingresa la cantidad de unidades a agregar: "))
        if producto in productos_stock:
            productos_stock[producto] += unidades
            print(f"Nuevo stock de {producto}: {productos_stock[producto]}")
        else:
            print(f"El producto {producto} no existe.")

    elif opcion == "3":
        print("\nAgregar nuevo producto:")
        producto = input("Ingresa el nombre del nuevo producto: ")
        if producto not in productos_stock:
            unidades = int(input("Ingresa la cantidad de unidades: "))
            productos_stock[producto] = unidades
            print(f"Producto {producto} agregado con stock: {unidades}")
        else:
            print(f"El producto {producto} ya existe.")

    elif opcion == "4":
        print("Saliendo...")
        break

    else:
        print("Opción inválida. Intenta nuevamente.")
```

```python
""" 
9) Creá una agenda donde las claves sean tuplas de (día, hora) y los valores sean eventos. 
Ejemplo: 
Permití consultar qué actividad hay en cierto dia ingresados por el usuario.
"""

fecha = {('Lunes', '10:00'): 'Reunión de equipo',
         ('Martes', '14:00'): 'Clase de programación',
         ('Miércoles', '09:00'): 'Cita médica',
         ('Jueves', '11:00'): 'Reunión de ventas',
         ('Viernes', '16:00'): 'Evento social'
}

print("Seleccione el dia que desea consultar:")
dia = input("Ingrese el día (Lunes, Martes, Miércoles, Jueves, Viernes): ")

key = fecha.keys()

if dia in [k[0] for k in key]:
    for k in key:
        if dia == k[0]:
            print(f"El {k[0]} a las {k[1]} tiene: {fecha[k]}")
else:
    print("No hay actividades programadas para ese día.")
```

```python
""" 
10) Dado un diccionario que mapea nombres de países con sus capitales, construí un nuevo 
diccionario donde: 
• Las capitales sean las claves. 
• Los países sean los valores.
"""

paises_capitales = {
    'Argentina': 'Buenos Aires',
    'Brasil': 'Brasilia',
    'Chile': 'Santiago',
    'Colombia': 'Bogotá',
    'Uruguay': 'Montevideo'
}

capitales_paises = {v: k for k, v in paises_capitales.items()}

print("Diccionario original (países -> capitales):")
for i in paises_capitales:
    print(f"{i} -> {paises_capitales[i]}")

print("\nDiccionario invertido (capitales -> países):")
for i in capitales_paises:
    print(f"{i} -> {capitales_paises[i]}")
```
