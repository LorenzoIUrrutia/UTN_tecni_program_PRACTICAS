```python
# 1 lista de 10 estudiantes, mostrar lista completa, ordenada
# y la nota mas alta y baja

notas: list[int] = []


nota_1 = int(input("Ingrese la primera nota: "))
while nota_1 < 0 or nota_1 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_1 = int(input("Ingrese la primera nota: "))
notas.append(nota_1)

nota_2 = int(input("Ingrese la segunda nota: "))
while nota_2 < 0 or nota_2 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_2 = int(input("Ingrese la segunda nota: "))
notas.append(nota_2)

nota_3 = int(input("Ingrese la tercera nota: "))
while nota_3 < 0 or nota_3 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_3 = int(input("Ingrese la tercera nota: "))
notas.append(nota_3)

nota_4 = int(input("Ingrese la cuarta nota: "))
while nota_4 < 0 or nota_4 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_4 = int(input("Ingrese la cuarta nota: "))
notas.append(nota_4)

nota_5 = int(input("Ingrese la quinta nota: "))
while nota_5 < 0 or nota_5 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_5 = int(input("Ingrese la quinta nota: "))
notas.append(nota_5)

nota_6 = int(input("Ingrese la sexta nota: "))
while nota_6 < 0 or nota_6 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_6 = int(input("Ingrese la sexta nota: "))
notas.append(nota_6)

nota_7 = int(input("Ingrese la septima nota: "))
while nota_7 < 0 or nota_7 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_7 = int(input("Ingrese la septima nota: "))
notas.append(nota_7)

nota_8 = int(input("Ingrese la octava nota: "))
while nota_8 < 0 or nota_8 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_8 = int(input("Ingrese la octava nota: "))
notas.append(nota_8)

nota_9 = int(input("Ingrese la novena nota: "))
while nota_9 < 0 or nota_9 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_9 = int(input("Ingrese la novena nota: "))
notas.append(nota_9)

nota_10 = int(input("Ingrese la decima nota: "))
while nota_10 < 0 or nota_10 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_10 = int(input("Ingrese la decima nota: "))
notas.append(nota_10)

print("Las notas ingresadas son: ", notas)

cantidad_notas = len(notas)
for i in range(cantidad_notas -1):
    for j in range(cantidad_notas - 1 - i):
        if notas[j] > notas[j + 1]:
            notas[j], notas[j + 1] = notas[j + 1], notas[j]

print("Las notas ordenadas son: ", notas)
print("La nota mas baja es: ", notas[0])
print("La nota mas alta es: ", notas[-1])
```

```python
# 2 lista de 5, mostrar alfabeticamente
# pedir eliminar uno y mostrar nuevamente

items: list[str] = []
item = ""

for i in range(5):
    item = input(f"Ingrese el item {i + 1}: ")
    items.append(item)


print(f"Items ingresados: {sorted(items)}")

print("Desea eliminar algun item? (s/n)")
respuesta = input().lower()

if respuesta == "s":
    item_eliminar = input("Ingrese el item a eliminar: ")
    if item_eliminar in items:
        items.remove(item_eliminar)
    print(f"Lista actualizada: {sorted(items)}")
else:
    print("No se elimino ningun item")
    
```

```python
# 5 lista con 8 estudiantes presentes
# preguntar para agregar otro estudiante o eliminar
# mostrar lista actualizada

estudiantes: list[str] = []

for i in range(8):
    print(f"Ingrese el nombre del estudiante {i + 1}: ")
    estudiante = input()
    estudiantes.append(estudiante)
    print(f"Ha ingresado al estudiante: {estudiante}")
    print()
print(f"""Estudiantes presentes 
{sorted(estudiantes)}""")

print("Desea agregar o eliminar un estudiante? (s/n)")
opcion = input()

if opcion == "s":
    print("Desea agregar (a) o eliminar (e) un estudiante?")
    accion = input()
    if accion == "a":
        print("Ingrese el nombre del nuevo estudiante: ")
        nuevo_estudiante = input()
        estudiantes.append(nuevo_estudiante)
        print(f"Lista actualizada: {sorted(estudiantes)}")
    elif accion == "e":
        print("Ingrese el nombre del estudiante a eliminar: ")
        estudiante_eliminar = input()
        if estudiante_eliminar in estudiantes:
            estudiantes.remove(estudiante_eliminar)
            print(f"Lista actualizada: {sorted(estudiantes)}")
elif opcion == "n":
    print("No se realizaron cambios en la lista de estudiantes")    
```

```python
# 6 lista con 7 numeros
# rotar todos a la derecha una posicion
# ultimo pasa a ser primero

list_num: list[int] = [10, 20, 30, 40, 50, 60, 70]
print(f"Lista original: {list_num}")

ultimo = list_num[-1]
for i in range(len(list_num) - 1, 0, -1):
    list_num[i] = list_num[i - 1]
list_num[0] = ultimo  # Poner 70 al principio

print(f"Lista rotada: {list_num}")
```

```python
# 7 matriz 7x2 temp minima y maxima de una semana
# promedio de maximas y minimas
# en que dia fue la de mayor amplitud

temperaturas: list[list[int]] = [[30, 20], [32, 22], [31, 21], [33, 23], [34, 24], [35, 25], [36, 26]]
print(f"""Temperaturas de la semana:
Dia 1: Max {temperaturas [0][0]}°C, Min {temperaturas[0][1]}°C
Dia 2: Max {temperaturas [1][0]}°C, Min {temperaturas[1][1]}°C
Dia 3: Max {temperaturas [2][0]}°C, Min {temperaturas[2][1]}°C
Dia 4: Max {temperaturas [3][0]}°C, Min {temperaturas[3][1]}°C
Dia 5: Max {temperaturas [4][0]}°C, Min {temperaturas[4][1]}°C
Dia 6: Max {temperaturas [5][0]}°C, Min {temperaturas[5][1]}°C
Dia 7: Max {temperaturas [6][0]}°C, Min {temperaturas[6][1]}°C
""")
suma_minimas = 0
suma_maximas = 0

for i in range(len(temperaturas)):
    suma_maximas += temperaturas[i][0]
    suma_minimas += temperaturas[i][1]
promedio_maximas = suma_maximas / len(temperaturas)
promedio_minimas = suma_minimas / len(temperaturas)
print(f"""Promedio de temperaturas máximas: {promedio_maximas}°C
Promedio de temperaturas mínimas: {promedio_minimas}°C
""")
mayor_amplitud = 0
dia_mayor_amplitud = 0 
for i in range(len(temperaturas)):
    amplitud = temperaturas[i][0] - temperaturas[i][1]
    if amplitud > mayor_amplitud:
        mayor_amplitud = amplitud
        dia_mayor_amplitud = i + 1
print(f"La mayor amplitud térmica fue de {mayor_amplitud}°C el día {dia_mayor_amplitud}")
```

```python
# 10 matrix 4 productos x 7 dias
# total vendido cada producto
# dia con mas ventas
# producto mas vendido

semanal: list[list[int]] = [[5, 3, 2, 7, 8, 10, 15], 
                      [6, 20, 12, 18, 25, 30, 35], 
                      [2, 6, 4, 9, 30, 40, 50], 
                      [4, 8, 6, 11, 20, 25, 30]]

#total por producto 
for i in range(len(semanal)):
    total_producto = sum(semanal[i])
    print(f"Total vendido del producto {i + 1}: {total_producto}")

#dia con mas ventas
total_dia: list[int] = [0, 0, 0, 0, 0, 0, 0]
for j in range(len(semanal[0])):
    for i in range(len(semanal)):
        total_dia[j] += semanal[i][j]
dia_mas_ventas = total_dia.index(max(total_dia)) + 1
print(f"Día con más ventas: {dia_mas_ventas}")

#producto mas vendido
totales_productos: list[int] = []
for i in range(len(semanal)):
    totales_productos.append(sum(semanal[i]))
producto_mas_vendido = totales_productos.index(max(totales_productos)) + 1
print(f"Producto más vendido: {producto_mas_vendido}")
```
