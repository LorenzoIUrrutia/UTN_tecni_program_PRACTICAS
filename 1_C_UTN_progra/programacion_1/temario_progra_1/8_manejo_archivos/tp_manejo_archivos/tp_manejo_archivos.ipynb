```python
# 3) Agregar productos desde teclado: Modificar el programa para que luego de mostrar 
# los productos, le pida al usuario que ingrese un nuevo producto (nombre, precio, 
# cantidad) y lo agregue al archivo sin borrar el contenido existente.

nuevo_producto = input("Ingrese el nombre del nuevo producto: ")
nuevo_precio = input("Ingrese el precio del nuevo producto: ")
nueva_cantidad = input("Ingrese la cantidad del nuevo producto: ")

with open("productos.csv", "a") as archivo_csv:
    archivo_csv.write(f"{nuevo_producto},{nuevo_precio},{nueva_cantidad}\n")
```

```python
# 4) Cargar productos en una lista de diccionarios: Al leer el archivo, cargar los datos en 
# una lista llamada productos, donde cada elemento sea un diccionario con claves: 
# nombre, precio, cantidad

productos_dict = []

with open("productos.csv", "r") as archivo_csv:
    archivo_csv.readline()  # Saltar la cabecera
    for linea in archivo_csv:
        nombre, precio, cantidad = linea.strip().split(",")
        producto = {
            "Nombre": nombre,
            "Precio": float(precio),
            "Cantidad": int(cantidad)
        }
        productos_dict.append(producto)

print(productos_dict)
```

```python
# 5. Buscar producto por nombre: Pedir al usuario que ingrese el nombre de un 
# producto. Recorrer la lista de productos y, si lo encuentra, mostrar todos sus datos. Si 
# no existe, mostrar un mensaje de error. 

print("\nBúsqueda de producto")
print("Lista de productos disponibles:")
for producto in productos_dict:
    print(f" - {producto['Nombre']}")

nombre_buscar = input("Ingrese el nombre del producto a buscar: ")
producto_encontrado = None
for producto in productos_dict:
    if producto["Nombre"] == nombre_buscar:
        producto_encontrado = producto
        break

if producto_encontrado:
    print("Producto encontrado:")
    for clave, valor in producto_encontrado.items():
        print(f"  {clave}: {valor}")
else:
    print("Producto no encontrado.")

```
