```python
# 1 lista de 10 estudiantes, mostrar lista completa, ordenada
# y la nota mas alta y baja

notas = []

nota_1 = int(input("Ingrese la primera nota: "))
while nota_1 < 0 or nota_1 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_1 = int(input("Ingrese la primera nota: "))
notas.append(nota_1)

nota_2 = int(input("Ingrese la segunda nota: "))
while nota_2 < 0 or nota_2 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_2 = int(input("Ingrese la segunda nota: "))
notas.append(nota_2)

nota_3 = int(input("Ingrese la tercera nota: "))
while nota_3 < 0 or nota_3 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_3 = int(input("Ingrese la tercera nota: "))
notas.append(nota_3)

nota_4 = int(input("Ingrese la cuarta nota: "))
while nota_4 < 0 or nota_4 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_4 = int(input("Ingrese la cuarta nota: "))
notas.append(nota_4)

nota_5 = int(input("Ingrese la quinta nota: "))
while nota_5 < 0 or nota_5 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_5 = int(input("Ingrese la quinta nota: "))
notas.append(nota_5)

nota_6 = int(input("Ingrese la sexta nota: "))
while nota_6 < 0 or nota_6 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_6 = int(input("Ingrese la sexta nota: "))
notas.append(nota_6)

nota_7 = int(input("Ingrese la septima nota: "))
while nota_7 < 0 or nota_7 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_7 = int(input("Ingrese la septima nota: "))
notas.append(nota_7)

nota_8 = int(input("Ingrese la octava nota: "))
while nota_8 < 0 or nota_8 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_8 = int(input("Ingrese la octava nota: "))
notas.append(nota_8)

nota_9 = int(input("Ingrese la novena nota: "))
while nota_9 < 0 or nota_9 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_9 = int(input("Ingrese la novena nota: "))
notas.append(nota_9)

nota_10 = int(input("Ingrese la decima nota: "))
while nota_10 < 0 or nota_10 > 10:
    print("Error, la nota debe estar entre 0 y 10")
    nota_10 = int(input("Ingrese la decima nota: "))
notas.append(nota_10)

print("Las notas ingresadas son: ", notas)

cantidad_notas = len(notas)
for i in range(cantidad_notas -1):
    for j in range(cantidad_notas - 1 - i):
        if notas[j] > notas[j + 1]:
            notas[j], notas[j + 1] = notas[j + 1], notas[j]

print("Las notas ordenadas son: ", notas)
print("La nota mas baja es: ", notas[0])
print("La nota mas alta es: ", notas[-1])
```
